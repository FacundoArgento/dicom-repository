#### Models

##### Data for the script

```python
# configure names for save the results
def config_params(weights_name, test=True):
  weights_path = f'/project/Models/Base/weights/{weights_name}.h5'
  data_path = test_dir if test else '/content/pre-process\ data'
  save_path = f'{weights_name}\ results'
  return weights_path, data_path, save_path
```

```python
# available names models for testing
import os
p = '/project/Models/Base/weights'
l = !ls {p}
models = [m.split('.')[0][1:] for m in l if os.path.isfile(f'{p}/{m[1:-1]}')]
for name in models: print(name)
```

```python
# copy one name from the above list here 
# (use \ for escape white spaces and special characters)
weights_name = 'unet_with_efficientnetb7\ \(8bs\)'
weights, data, save = config_params(weights_name)
```
